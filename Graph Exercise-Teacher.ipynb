## Make a bar graph with each cow's weight

#### <span style='color:red'> Task </span>
Make a graph plotting the weight of each cow observation.

```python
species = 'cow'

graphtype = 'bar'
graphtitle = f'Animal Weights by Observation ({species})'

figsize = (20, 10)

xlabel = 'Observation'
ylabel = 'Weight'

# Override the default font sizes for title and labels
title_fontsize = 18
label_fontsize = 15

ax = animals[species].plot(kind=graphtype, 
                           title=graphtitle,
                           xlabel=xlabel,
                           ylabel=ylabel,
                           figsize=figsize
                          )

ax.title.set_size(title_fontsize)
ax.xaxis.label.set_size(label_fontsize)
ax.yaxis.label.set_size(label_fontsize)
```

## Make a graph plotting the mean weight of each species

#### <span style='color:red'> Task </span>
Make a bar graph plotting the mean weight of each species. 

Hint: Use `mean()` to calculate the mean.

```python
graphtype = 'bar'
graphtitle = 'Mean of Animal Weights by Species'

figsize = (15, 7)

xlabel = 'Species'
ylabel = 'Weight'

ax = animals.mean().plot(kind=graphtype, 
                         title=graphtitle,
                         xlabel=xlabel,
                         ylabel=ylabel,
                         figsize=figsize
                        )
```

## Make a histogram of the weight distribution (bins) of each species

#### <span style='color:red'> Task </span>
Make a histogram of the weight distribution of each species.

```python
graphtype = 'hist'
graphtitle = 'Animal Weight Distribution'

figsize = (15, 7)

xlabel = 'Weight'
ylabel = 'Frequency'

bins = 20

ax = animals.plot(kind=graphtype, 
                  title=graphtitle,
                  edgecolor='w',
                  bins=bins,
                  xlabel=xlabel,
                  ylabel=ylabel,
                  figsize=figsize
                 )

result = ax.set_xlabel(xlabel)
result = ax.set_ylabel(ylabel)
```

## Examples using `melt` and `pivot`

```python
animals_tidy
```

```python
animals_pivot = animals_tidy.pivot(columns='species', values='weight').rename_axis(None, axis='columns')
```

```python
animals_pivot
```

```python
graphtype = 'box'
graphtitle = 'Animal Weights Box Plot'

figsize = (15, 7)

xlabel = 'Species'
ylabel = 'Weight'

ax = animals_pivot.plot(kind=graphtype, 
                        title=graphtitle,
                        xlabel=xlabel,
                        ylabel=ylabel,
                        figsize=figsize
                       )
```

```python
animals_tidy_new = animals_pivot.melt(var_name='species', value_name='weight').dropna().reset_index()
```
