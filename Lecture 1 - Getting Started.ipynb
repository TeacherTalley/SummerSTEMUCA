#### Std (Standard deviation)
A standard deviation is a measure of how dispersed the data is in relation to the mean. Low standard deviation means data are clustered around the mean, and high standard deviation indicates data are more spread out.

````python
traffic.std()
````

Calculate a set of basic statistics by street (column).

````python
traffic.describe()
````

What does this data mean? 

Quartiles are three values that split sorted data into four parts, each with an equal number of observations. 

- First quartile (25%): Also known as Q1, or the lower quartile. This is the number halfway between the lowest number and the middle number.
- Second quartile (50%) : Also known as Q2, or the median. This is the middle number halfway between the lowest number and the highest number.
- Third quartile (75%): Also known as Q3, or the upper quartile. This is the number halfway between the middle number and the highest number.


Calculate a set of basic statistics for LP410 street.

````python
traffic['LP410'].describe()
````

#### Sort Data

````python
street = 'LP410'
````

Sort the data from lightest traffic to heaviest.

````python
traffic[street].sort_values()
````

Sort from highest to lightest traffic.

````python
traffic[street].sort_values(ascending=False)
````

Alternatively, you can sort the entire dataframe by a column name using the `by` parameter.

````python
traffic.sort_values(by=street, ascending=False)
````
