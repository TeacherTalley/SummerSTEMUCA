## Graphing Data

#### Histograms
A histogram is a graph used to represent the frequency distribution of one variable. Histograms often classify data into various bins or range groups and count how many data points belong to each of those bins.

Create a simple histogram of the beak sizes. We will use the plot method on the panda dataframe to create the histogram.

```python
ax = beaks.plot(kind='hist')
```

Let's add a title to our graph.

```python
graphtype = 'hist'
graphtitle = f'Daphne Island Finches (n = {observations})'

ax = beaks.plot(kind=graphtype, 
                title=graphtitle
               )
```

We can add labels for the x and y axis.

```python
xlabel = 'Beak size in mm'
ylabel = 'Number of birds'

ax = beaks.plot(kind=graphtype, 
                title=graphtitle
               )

# Augment our graph
#   Note: Pandas before version 2.0 does not support xlabel and ylabel on histograms.  All other chart types support this on the call.
result = ax.set_xlabel(xlabel)
result = ax.set_ylabel(ylabel)
```

It is a little hard to see the separation between values. Let's change the bars to green, add an edge color of white, and a grid to better show values.

```python
xlabel = 'Beak size in mm'
ylabel = 'Number of birds'

ax = beaks.plot(kind=graphtype, 
                title=graphtitle,
                color='g',
                edgecolor='w',
                grid=True
               )

# Augment our graph
result = ax.set_xlabel(xlabel)
result = ax.set_ylabel(ylabel)
```

The default number of bins to count continuous values is 10, but we can change that.

```python
xlabel = 'Beak size in mm'
ylabel = 'Number of birds'

bins = 20

ax = beaks.plot(kind=graphtype, 
                title=graphtitle,
                color='g',
                edgecolor='w',
                bins=bins,
                grid=True
               )

# Augment our graph
result = ax.set_xlabel(xlabel)
result = ax.set_ylabel(ylabel)
```

Let's make our plot size a little bit bigger.

```python
xlabel = 'Beak size in mm'
ylabel = 'Number of birds'

figsize = (15, 7)

ax = beaks.plot(kind=graphtype, 
                title=graphtitle,
                color='g',
                edgecolor='w',
                grid=True,
                figsize=figsize
               )

# Augment our graph
result = ax.set_xlabel(xlabel)
result = ax.set_ylabel(ylabel)
```

### Bar Graphs
A bar graph uses bars at different heights to depict values of the variables in an observation.

Note that bar charts plot the actual value of a variable rather than frequency (count) of the value plotted by histograms.

Let's use traffic data to make a bar chart.

#### Traffic Volumes

#### Data Source

- ***TrafficCount.csv*** - Traffic count on Daphne Island streets by hour

```python
traffic_datafile = 'TrafficCount.csv'
```

```python
traffic = pd.read_csv(traffic_datafile)
```

```python
traffic
```

```python
graphtype = 'bar'
graphtitle = 'Traffic on Three Streets'

figsize = (10, 5)

xlabel = 'Hour'
ylabel = 'Car Count'

ax = traffic.plot(kind=graphtype, 
                  title=graphtitle,
                  grid=True,
                  xlabel=xlabel,
                  ylabel=ylabel,
                  figsize=figsize
                 )
```

Change the graph to a stacked bar chart.

```python
graphtype = 'bar'
graphtitle = 'Traffic on Three Streets'

figsize = (10, 5)

xlabel = 'Hour'
ylabel = 'Car Count'

ax = traffic.plot(kind=graphtype, 
                  title=graphtitle,
                  grid=True,
                  stacked=True,
                  xlabel=xlabel,
                  ylabel=ylabel,
                  figsize=figsize
                 )
```

Another type of graph is a horizontal bar chart or `barh`.

Don't forget to reverse the x and y labels...

```python
graphtype = 'barh'
graphtitle = 'Traffic on Three Streets'

figsize = (15, 7)

xlabel = 'Car Count'
ylabel = 'Hour'

ax = traffic.plot(kind=graphtype, 
                  title=graphtitle,
                  grid=True,
                  xlabel=xlabel,
                  ylabel=ylabel,
                  figsize=figsize
                 )
```

We can also create a stacked bar chart by adding the `stacked` parameter.

```python
graphtype = 'barh'
graphtitle = 'Traffic on Three Streets'

figsize = (15, 7)

xlabel = 'Car Count'
ylabel = 'Hour'

ax = traffic.plot(kind=graphtype, 
                  title=graphtitle,
                  grid=True,
                  stacked=True,
                  xlabel=xlabel,
                  ylabel=ylabel,
                  figsize=figsize
                 )
```

### Box Plots
A boxplot is a standardized way of displaying the distribution of data .

Box plots can show multiple statistics about a variable in a single chart. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be outliers using a method that is a function of the inter-quartile range.

The diagram below giving a description of a box plot came from this site: https://chartio.com/learn/charts/box-plot-complete-guide/

```python
graphtype = 'box'
graphtitle = 'Monthly Measles Cases NYC: 1931-1971'

figsize = (15, 15)

xlabel = 'Month'
ylabel = 'Number of cases'

ax = measles.plot(kind=graphtype, 
                  title=graphtitle,
                  grid=True,
                  xlabel=xlabel,
                  ylabel=ylabel,
                  figsize=figsize
                 )
```

### Pie Charts

Let's show the totals in a pie chart.  First, we need to sum the values by month.

```python
measles.sum()
```

Set the graph parameters...

```python
graphtype = 'pie'
graphtitle = f'Monthly Measles Cases for NYC for 1931-1971'

figsize = (10, 10)

labelfontsize = 7

ax = measles.sum().plot(kind=graphtype, 
                        title=graphtitle,
                        fontsize=labelfontsize,
                        figsize=figsize
                       )
```

You can represent the same information in the pie chart above with a bar chart.

```python
graphtype='bar'
graphtitle = f'Monthly Measles Cases for NYC for 1931-1971'

figsize = (15, 7)

xlabel = 'Month'
ylabel = 'Number of cases'

ax = measles.sum().plot(kind=graphtype, 
                        title=graphtitle,
                        grid=True,
                        xlabel=xlabel,
                        ylabel=ylabel,
                        figsize=figsize
                       )
```

Now we can sum across the months for each year.

The `axis='columns'` parameter on sum causes sums to be computed across rows rather than down columns.

```python
measles.sum(axis='columns')
```

```python
graphtype='bar'
graphtitle = f'Yearly Measles Cases for NYC for 1931-1971'

figsize = (15, 7)

xlabel = 'Year'
ylabel = 'Number of cases'

ax = measles.sum(axis='columns').plot(kind=graphtype, 
                                      title=graphtitle,
                                      grid=True,
                                      xlabel=xlabel,
                                      ylabel=ylabel,
                                      figsize=figsize
                                     )
```

### Scatter Plots

### Data Source
- ***DaphneIsland.csv*** - Beak sizes for child and parents of Daphne Island finches in mm

NOTE: This is a different set of data from the earlier beak data.

```python
beak_family_datafile = 'DaphneIsland.csv'
```

```python
beak_family = pd.read_csv(beak_family_datafile)
beak_family 
```

Let's graph that data for the child and parent 1 first.

```python
graphtype = 'scatter'
graphtitle = 'Child by Parent 1 Beak Size'

figsize = (15, 7)

xlabel = 'Parent'
ylabel = 'Child'

# set the marker size and color
size1 = 40
color1 = 'green'

ax = beak_family.plot(kind=graphtype, 
                      title=graphtitle,
                      x='Parent 1 Beak Size',
                      y='Child Beak Size',
                      color=color1,
                      s=size1,
                      xlabel=xlabel,
                      ylabel=ylabel,
                      figsize=figsize
                     )
```

We can superimpose two scatter plots on the same axes.  We will put both parents on the same chart with different colors. 

Just for illustration, change the size and color of the data for each parent.

```python
graphtype = 'scatter'
graphtitle = 'Child by Both Parents Beak Size'

figsize = (15, 7)

xlabel = 'Parent'
ylabel = 'Child'

# set the marker size and color
size1 = 40
color1 = 'green'
size2 = 60
color2 = 'red'

ax = beak_family.plot(kind=graphtype, 
                      title=graphtitle,
                      x='Parent 1 Beak Size',
                      y='Child Beak Size',
                      color=color1,
                      s=size1,
                      label='Parent 1',
                      figsize=figsize
                     )

# Note the axes(ax) created above is referenced when calling this plot
ax = beak_family.plot(kind=graphtype, 
                      x='Parent 2 Beak Size',
                      y='Child Beak Size',
                      color=color2,
                      s=size2,
                      label='Parent 2',
                      xlabel=xlabel,
                      ylabel=ylabel,
                      ax=ax
                     )
```

## Data Manipulation

Our source data is not always in a convenient format for graphing.  We are going to learn a few `pandas`  tricks to manipulate the data into the format we need for our graphs.

We have done some of this already in previous examples.  Let's look at the measles data again.

```python
measles
```

Recall that we used `sum` to sum the data in each column (or row) to build the chart that we wanted.   We can also save the results of the sum into a variable that we can use later. 

```python
measles_sum_columns = measles.sum()
measles_sum_columns
```

```python
measles_sum_rows = measles.sum(axis='columns')
measles_sum_rows
```

You can use these newly created tables to build charts.

### Selecting specific rows and columns.

You can use `loc` to select rows or columns.  The format is table.loc[rows,columns].  
* The colon `:` is use to specify what split of the data we want.  
* Notice that year is an integer, so we don't put the value in quotes.  

Let's select just the row for 1931 and all columns.

```python
measles.loc[1931,:]
```

Now select 1940 - 1949.  This time we will save the results to a variable to use later in a graph.

```python
measles_40s = measles.loc[1940:1949,:]
measles_40s 
```

Now select all years, but only the months January through March.
* Notice that month is a string, so we put quotes around the column names.

```python
measles.loc[:,'January':'February']
```

Let's use what we learned to make a graph that only shows the total measles cases by month for the 1940's

```python
graphtype='bar'
graphtitle = 'Measles Cases By Year NYC: 1940-1949'

figsize = (15, 7)

xlabel = 'Year'
ylabel = 'Cases'

ax = measles_40s.sum().plot(kind=graphtype, 
                            title=graphtitle,
                            grid=True,    
                            xlabel=xlabel,
                            ylabel=ylabel,
                            figsize=figsize
                           )
```
